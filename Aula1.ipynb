```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo os parâmetros da matriz
linhas = 40
colunas = 40
valor_max = 255

# Gerando a matriz
matriz = np.zeros((linhas, colunas))

for i in range(linhas):
    for j in range(colunas):
        matriz[i][j] = ((i + j) % valor_max)  # Definindo os valores da matriz

# Plotando a imagem em tons de cinza
plt.imshow(matriz, cmap='gray')
plt.show()
```
